```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize
```

```python
data = {}

data.update({4.7:818})
data.update({3.7:777})
data.update({3.0:700})
data.update({2.2:640})
data.update({2.0:585})

data.update({1.5:540})
data.update({1.23:470})

data.update({0.82:360})

data.update({0.47:242})
data.update({0.37:200})
data.update({0.24:132})
data.update({0.15:83})

data.update({0.1:56})
```

```python
x,y= np.array(list(data.keys())),np.array(list(data.values()))
plt.plot(x,y,'o')
plt.show()
```

```python
y
```

```python
def fit_func (x,I0,p0,a,c):
    return I0*x/(x+p0) + a*x +c

def noise_func (x,a,c):
    return a*x +c

params, params_covariance = optimize.curve_fit(fit_func, x, y)#,p0=[],maxfev =20000)#,bounds = bounds)
```

```python
type(x)
```

```python
print(params)
plt.plot(x,y,'o',color='black',label = 'data')
plt.plot(x, fit_func(x,*params),'r-',label = 'fit')

plt.plot(x, noise_func(x,params[-2],params[-1]),label = 'noise',color = 'black')
title = 'Counts in saturation ' + str(np.round(params[0],2))+ ' kcounts/s \n'
plt.title(title, fontsize = 20)
plt.xlabel('Power, mW', fontsize = 15)
plt.ylabel('kcounts/sec', fontsize = 15)
plt.legend()
plt.savefig('sat curve 27.02.png')
plt.show()
```
