```python
it = np.arange(10,1000,10)
data_0 = trace_0_400 / trace_0_400[3000]
data_1 = trace_1_400 / trace_1_400[3000]

cumsum_0 = []
cumsum_1 = []

for i in it:
    
    cumsum_0.append(np.sum(data_0[0:int(i)]))
    cumsum_1.append(np.sum(data_1[0:int(i)]))
plt.plot(it,cumsum_0,color = 'red')
plt.plot(it,cumsum_1,color = 'black')
plt.show()
```

```python
snr = np.sqrt(np.array(cumsum_0)+ np.array(cumsum_1)) * (np.array(cumsum_0) - np.array(cumsum_1)) / np.array(cumsum_0)

# plt.plot(it,(np.array(cumsum_0) - np.array(cumsum_1)) / np.array(cumsum_0))
plt.plot(it,snr)

plt.show()
```
