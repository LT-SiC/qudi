```python
from matplotlib import pyplot as plt
```

```python
y1 = lambda B: np.abs(2870-2.8*B)
y2 = lambda B: np.abs(2870+2.8*B)
```

```python
plt.figure(dpi=150)
B = linspace(0,2000,1000)
plt.plot(B,y1(B))
plt.plot(B,y2(B))
plt.fill_between([B[0],B[-1]],[1000,1000],[3300,3300],color="r")
plt.fill_between([B[0],B[-1]],[1000,1000],[3300,3300])
plt.show()
```

```python
rabi_period = np.array([100000000,1122,397,228,178,160,150,141,134])*1e-9
awg_ampl = np.array([0,0.1,0.3,0.5,0.6,0.7,0.8,0.9,0.99])

rabi_period_pa = np.array([115,116,118,129.5,185,396,1000000000000])*1e-9
awg_ampl_pa = np.array([1,0.8,0.6,0.4,0.2,0.1,0])

rabi_period_pa_att = np.array([1000000,687,293,159,137,125,122])*1e-9
awg_ampl_pa_att=np.array([0,0.1,0.3,0.5,0.7,0.9,1.0])


plt.plot(awg_ampl_pa_att,1/rabi_period_pa_att/1e6,'o-',color='red',label = 'with PreAmp 32dDb attenuation')

plt.plot(awg_ampl_pa,1/rabi_period_pa/1e6,'o-',color='black',label = 'with PreAmp 26dB attenuation')
plt.plot(sorted(awg_ampl),sorted(1/rabi_period/1e6),'o-',label = 'No PreAmp')
plt.ylabel('Rabi Freq, MHz')
plt.xlabel('AWG amplitude')
plt.ylim(0,10)
plt.legend()
# plt.plot(awg_ampl,awg_ampl*9)
# plt.plot(awg_ampl,awg_ampl*9/1.4)
plt.grid()
plt.show()
```
