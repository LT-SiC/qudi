```python
names = ['time','signal','aa','bb']
df= pd.read_csv(file,skiprows = 18, delimiter= '\t',names = names)
del df['aa']
del df['bb']
df.head()
```

```python
max(df['signal'])
```

```python
xdata,ydata = df['time'],df['signal']
ydata=ydata-np.mean(ydata)
plt.plot(xdata*1e6,ydata)
plt.show()
```

```python
x =[]
y=[]
for i in range(len(xdata.values)):
    if xdata[i]<5.1e-6:
        x.append(xdata[i]*1e6)
        y.append(ydata[i])
x=np.array(x)
y=np.array(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# def fit_func (x,a,b,c,t):
#     return a * np.cos(b * x)*np.exp(-x/t) + c

def fit_func (x,a,c,t):
    return a * np.exp(-x/t) + c
```

```python
# bounds = ([0,-0.5,0],[10,0.5,10])


# x=x[::2]
# y=y[::2]
params, params_covariance = optimize.curve_fit(fit_func, x, y)#,p0=[-1,1,0,1],maxfev =5000)#,bounds = bounds)
print(params)
plt.plot(x,y)
plt.plot(x,fit_func(x, *params))
plt.show()
```
